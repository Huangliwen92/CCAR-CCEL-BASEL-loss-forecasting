# Build vintages

This example notebook constructs vintage cohorts from the sample data and produces a vintage-by-age pivot table of balances by delinquency bucket.

```
import pandas as pd

# Adjust the path if running from the repo root or the notebooks folder
df = pd.read_csv('../data/sample_vintage_data.csv', parse_dates=['origination_date','observation_date'])

# compute age in months
df['age_months'] = (df['observation_date'].dt.to_period('M') - df['origination_date'].dt.to_period('M')).apply(lambda x: x.n)

# vintage table: sum of balances by vintage, age, and delinquency state
vintage = df.groupby(['origination_date','age_months','delinquency_bucket'])['balance'].sum().reset_index()
pivot = vintage.pivot_table(index=['origination_date','age_months'], columns='delinquency_bucket', values='balance', fill_value=0)

# show result
print(pivot.head())
```
